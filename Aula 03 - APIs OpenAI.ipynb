###Autenticação da API

```python
from openai import OpenAI


client = OpenAI(api_key=os.environ.get('OPENAI_API_KEY'))
```

###API do ChatGPT

```python
response = client.chat.completions.create(
  model="gpt-3.5-turbo",  # Procure na documentação e teste com outros modelos
  messages=[
      {"role": "system", "content": "Você é um professor de programação especializado em inteligência artificial"},
      {"role": "user", "content": "Explique o conceito de inteligência artificial de forma simples para desenvolvedores."}
      ],
  max_tokens=150

)

print(response)
```

```python
response.choices[0].message.content
```

###API do DALL-E

```python
# Criando uma imagem a partir de uma descrição textual
response = client.images.generate(
  model="dall-e-2",  # Certifique-se de usar o modelo correto para DALL-E
  prompt="uma ilustração de um gato astronauta no espaço, estilo cartoon",
  n=1,  # Número de imagens a gerar
  size="1024x1024"  # Tamanho da imagem
)

# A resposta inclui um URL para a imagem gerada
print(response.data[0].url)
```

```python
response.data
```

###API do GPT para geração de Código

```python
# Gerando código a partir de um prompt
response = client.chat.completions.create(
  model="gpt-3.5-turbo",  # Procure na documentação e teste com outros modelos
  messages=[
      {"role": "system", "content": "Você é um professor de programação especializado em Python."}, #Retorne apenas o conteúdo do código gerado
      {"role": "user", "content": "Escreva uma função em Python que aceite um número inteiro e retorne True se o número for primo, caso contrário, retorne False."}
      ],
  max_tokens=1000,
  temperature=0.5  # Um valor mais baixo faz com que o modelo gere respostas mais determinísticas e focadas
)

print(response.choices[0].message.content.strip())
```

###API do Whisper

```python
# Caminho para o arquivo de áudio que você deseja transcrever
audio_file_path = 'Gravação.mp3'

audio_file= open(audio_file_path, "rb")
transcription = client.audio.transcriptions.create(
  model="whisper-1", 
  file=audio_file
)
print(transcription.text)
```

```python
#Tradução

audio_file= open(audio_file_path, "rb")
translation = client.audio.translations.create(
  model="whisper-1", 
  file=audio_file
  
)
print(translation.text)
```

###API do Embeddings

```python
# Textos para os quais você deseja gerar embeddings
texts = ["Aprendizado de máquina é o estudo de algoritmos de computador que melhoram automaticamente com a experiência.",
         "Inteligência artificial é a ciência de fazer as máquinas fazerem coisas que exigiriam inteligência se feitas por humanos."]

# Gerando embeddings para cada texto
embeddings = []
for text in texts:
    response = client.embeddings.create(
        input=text,
        model="text-embedding-3-small"  # Substitua pelo modelo de embeddings disponível
    )
    embeddings.append(print(response.data[0].embedding))

# Agora você tem uma lista de vetores de embeddings para cada texto
embeddings
```
